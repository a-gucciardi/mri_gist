```python
import ants

# Paths
moving_image_path = "sub-CC00058XX09_ses-11300_T2w_synthseg_robust_cpu.nii.gz"
template_image_path = "/Users/arnaud/Documents/GitHub/dhcp-volumetric-atlas-groupwise/mean/ga_44/template_t2.nii.gz"
output_path = "output_registered.nii.gz"

moving_image = ants.image_read(moving_image_path)
template_image = ants.image_read(template_image_path)

# Perform registration using SyN
registration = ants.registration(fixed=template_image, moving=moving_image, type_of_transform='SyN')

# Save the warped (registered) image
ants.image_write(registration['warpedmovout'], output_path)

print(f"Registered image saved to {output_path}")
```
