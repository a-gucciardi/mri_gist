```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

mean = 50
var = 50
sigma = var ** 0.5

image = Image.open('mri_image.png')
image = np.asarray(image)
noisy_image = np.copy(image)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        noisy_image[i][j] += np.random.normal(mean, sigma, (1,1))

figure, a = plt.subplots(1,2)
a[0].imshow(image, cmap='gray')
a[0].set_title('original image')
a[1].imshow(noisy_image, cmap='gray')
a[1].set_title('noisy image')
plt.show()
```

```python
from PIL import Image
import numpy as np

# Parameters for noise
mean = 50
var = 50
sigma = var ** 0.5

# Open the image
image = Image.open('mri_image.png').convert('L')  # Convert to grayscale
image_array = np.array(image)

# Create a noisy version
noisy_image_array = np.copy(image_array)
noise = np.random.normal(mean, sigma, image_array.shape)
noisy_image_array = np.clip(noisy_image_array + noise, 0, 255).astype(np.uint8)

# Convert back to PIL Image
noisy_image = Image.fromarray(noisy_image_array)

# Create a new image with both original and noisy side by side
result_width = image.width * 2
result_height = image.height
result = Image.new('L', (result_width, result_height))

# Paste the original and noisy images
result.paste(image, (0, 0))
result.paste(noisy_image, (image.width, 0))

# Save or show the result
result.save('result.jpg')
# Alternatively, to display:
# result.show()
```

```python
from PIL import Image, ImageDraw
import numpy as np

def create_irregular_mask(size):
    # Create a blank image for the mask
    mask = Image.new('L', size, 0)
    draw = ImageDraw.Draw(mask)
    
    # Draw an irregular shape
    points = [
        (50, 50), (100, 20), (150, 80), (200, 50),
        (250, 100), (200, 150), (150, 180), (100, 150),
        (50, 200), (20, 150), (50, 50)
    ]
    draw.polygon(points, fill=255)
    
    return mask

def create_image_with_shapes(size):
    # Create a blank image
    image = Image.new('RGB', size, (255, 255, 255))
    draw = ImageDraw.Draw(image)
    
    # Draw some shapes
    draw.ellipse([50, 50, 150, 150], fill=(255, 0, 0))  # Red circle
    draw.rectangle([180, 80, 280, 180], fill=(0, 255, 0))  # Green square
    draw.polygon([(100, 200), (150, 250), (50, 250)], fill=(0, 0, 255))  # Blue triangle
    
    return image

def apply_mask(image, mask):
    # Convert images to numpy arrays
    img_array = np.array(image)
    mask_array = np.array(mask)
    
    # Expand mask to 3 channels
    mask_array = np.expand_dims(mask_array, axis=2)
    mask_array = np.repeat(mask_array, 3, axis=2)
    
    # Apply mask
    result = np.where(mask_array == 255, img_array, 255)
    
    return Image.fromarray(result.astype('uint8'))

# Set image size
size = (300, 300)

# Create mask and image with shapes
mask = create_irregular_mask(size)
image = create_image_with_shapes(size)

# Apply mask to image
result = apply_mask(image, mask)

# Save results
mask.save('mask.png')
image.save('original.png')
result.save('result.png')

print("Images saved: mask.png, original.png, result.png")
```
