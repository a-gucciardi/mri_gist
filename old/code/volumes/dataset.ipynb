```python
dhcp_datadir = "/Users/arnaud/Downloads/sourcedata_orig"
dhcp_uniques = natsorted(glob(f"{dhcp_datadir}/sub*/ses*/anat/*.*"))
# 1050 : 492 (T1) + 558 (T2)
len(dhcp_uniques)
```

```python
participants = pd.read_csv("/Users/arnaud/Downloads/sourcedata_orig/participants.tsv", sep="\t")
participants.head(3)
```

```python
DATASET = "dhcp"
rows = []
for i, scan in tqdm(enumerate(dhcp_uniques[:10]), total=len(dhcp_uniques)):
    parts = scan.split("/")
    subject_id = parts[-4] # sub-XXXX
    session_id = parts[-3] # ses-XX
    img = parts[-1]

    gender = participants[participants["participant_id"] == f"{subject_id[4:]}"]['gender'].item(),
    age = participants[participants["participant_id"] == f"{subject_id[4:]}"]['birth_age'].item()

    dict1 = {"dataset": DATASET, 
             "mode": "T1" if "T1" in img else "T2", 
             "id": subject_id[4:], # remove sub- prefix
             "ses": session_id[4:],# remove ses- prefix
             "gender": gender[0],
             "gest_age": f"{age:.1f}",
             "source_file": scan,
             "mni_file": None,
             "aseg1": None,
             "qc": None,
             "vol": None
    }

    rows.append(dict1)

df = pd.DataFrame(rows)
df 
```
